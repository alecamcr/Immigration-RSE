All modules needed for the implementation of the project are imported. The module `working_functions.py` contains important self written functions for the preprocessing of the datasets. The modules `external_migration.py`, `internal_migration.py` and `sex_age.py` are independent programs upon which, the main program `migration_summary.py` is built up.

```python
import sex_age
import external_migration
import internal_migration
import working_functions as wf
```

The main program plots different figures containing important elements of the migration to Germany such as number of immigrants, origin countries, sex, age, and concentration and movement between Bundesländer. When calling `migration_summary.py` from terminal it'll plot by default 6 years before 2021, 5 countries from which the most migrants come, 5 Bundesländer to which the most migrants go, female and male with the following ages: 10, 20, 30, 40 and 50 years. Not all the data sets contain all (or the same information), so not all of them will be affected by the same arguments/information. Here some examples:

# 1. Data sets

For this project, we used three data sets from the page of the database of the Federal Statistical Office of Germany [destatis](https://www-genesis.destatis.de/genesis/online/data?operation=sprachwechsel&language=en), containing information about [sex and age](https://www-genesis.destatis.de/genesis//online?operation=table&code=12711-0006&bypass=true&levelindex=1&levelid=1685788618051#abreadcrumb), [migration between germany and foreign countries](https://www-genesis.destatis.de/genesis//online?operation=table&code=12711-0008&bypass=true&levelindex=1&levelid=1685788618051#abreadcrumb) (external migration) and [migration between bundesländer](https://www-genesis.destatis.de/genesis//online?operation=table&code=12711-0022&bypass=true&levelindex=1&levelid=1685788618051#abreadcrumb) (internal migration). First, we are going to recover the data sets. We get relative paths for the files.

```python
sex_age_data = wf.get_relative_path('data/12711-0006_sex_age.csv')
external_data = wf.get_relative_path('data/12711-0008_external_migration.csv')
internal_data = wf.get_relative_path('data/12711-0022_internal_migration.csv')
```

# 2. Migration summary

The `migration_summary.py` program, which can be run from the terminal, consists of the following functions, which draw different plots from file paths. They are:

# 2.1 Sex and age

## 2.1.1.

This program plots a barplot with sex and age. By default, it differentiates sex in the last 6 years from 2021 for following ages: 10,20,30,40 and 50. The only obligatory argument is the file path from which it takes the information to plot

```python
sex_age.draw_bar_plot(sex_age_data)
```

## 2.1.2.

Arguments can also be changed for a desired output. It can plot more or less year, more or less countries, ages and specific sex. If we want to display 9 years with children between 5 and 10 years we use the following arguments:

> Note: the maximum years to be plotted is 21

```python
sex_age.draw_bar_plot(sex_age_data, years=9, age=[5,6,7,8,9,10])
```

## 2.1.3.

If we want to ouput, additionaly, only male kids, we use the argument sex:

```python
sex_age.draw_bar_plot(sex_age_data, years=9, sex='male', age=[5,6,7,8,9,10])
```

# 2.2. External migration

This program plots a bar and a line plots countries and years. By default, it takes the 5 countries with most migrants in the last 6 years from 2021. The only obligatory argument is the file path from which it takes the information to plot.

## Line and bar plots

## 2.2.1.

Bar and line plots of most 5 migrant world countries coming to Germany in the last 6 years from 2021

```python
external_migration.draw_line_plot(external_data)
```

```python
external_migration.draw_bar_plot(external_data)
```

## 2.2.2.

It is possible to specify number of years `years` and/or number of countries `numcountries`. if we want the last 10 years with 9 countries:

```python
external_migration.draw_line_plot(external_data, years=10, numcountries=9)
```

```python
external_migration.draw_bar_plot(external_data, years=10, numcountries=9)
```

## 2.2.3.

If we are interested in plotting one country we can specify it with the argument `country`. The following program shows migration from Sweden in the past 10 years. 

> Note that `numcountries` and `country` are incompatible arguments. If both are given, it'll take only the country given

```python
external_migration.draw_line_plot(external_data, years=10, country='Sweden')
```

```python
external_migration.draw_bar_plot(external_data, years=10, country='Sweden')
```

# 2.3. Internal migration

## 2.3.1.

This program plots a scatter plot with the 5 Bundesländer receiving the most migration in the 6 past years from 2021 differentiating sex

```python
internal_migration.draw_scatter_plot(internal_data)
```

## 2.3.2.

It is also possible to specify years, number of Bundesländer, specific Bundesland and sex. The following program plots migration in the past 3 years and the 3 Bundesländer receiving the most migration

```python
internal_migration.draw_scatter_plot(internal_data, years=3, numcountries=3)
```

## 2.3.3.

If we want same as before but only males:

```python
internal_migration.draw_scatter_plot(internal_data, years=3, numcountries=3, sex='male')
```

## 2.3.4.

If, instead, we want to observe only females coming to Brandenburg in the same 3 years

```python
internal_migration.draw_scatter_plot(internal_data, years=3, country='Brandenburg', sex='female')
```
